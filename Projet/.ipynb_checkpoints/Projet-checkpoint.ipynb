```python
#Ce projet consiste à analyser des données concernant certains facteurs pouvant influencer le bohneur dans un groupe d'individu
#comme par exemple la situation professionnelle ou le salaire.
#Ce dataset ce concentre sur un groupe d'individus venant des Etats-unis

#importons les modules dont nous avons besoin
import numpy as np  
import pandas as pd #Il servira à manipuler notre dataset
import seaborn as sns
import matplotlib as mpl
from matplotlib import pyplot as plt #Il servira à faire des traitements mathématiques et à construire des graphes
# en fonction des resultats obtenus


df = pd.read_csv("Kag_happiness_indicators.csv",sep=",")
```

```python
df.columns
#Regardons quels colonnes sont disponibles dans ce dataset en appellant l'attribut columns du dataset
```

```python
data = df.filter(items=['year', 'workstat', 'prestige', 'divorce', 'widowed','educ','babies', 'preteen', 'teens','income', 'region','happy', 'owngun', 'tvhours'])
#Prenons les colonnes qui nous intéresse puis regardons le contenu en prenant un echantillon de 10 personnes
data
```

```python
#Voyons maintenant la distribution des valeurs de la colonne "happy" à travers 9 régions des Etats-unies
```

```python
vh = data[(data['happy']=='very happy')] 
#Prenons tous les valeurs dont la valeur de la colonne happy est "verry happy"

x = vh.groupby('region').count()['happy'] 
#Prenons les resultats précedents est regroupons les par région

plt.figure(figsize=(9,6))
#modifions la hauteur et largeur du graphe

plt.pie(x,autopct = lambda x: str(round(x, 2)) + '%', labels = ['Nord-est central', 'Sud-est central', 'Milieu Atlantique', 'Montagnes', 'Nouvelle-Anglettere', 'Pacifique', 'Atlantique-sud', 'Nord-est central', 'Sud-est central'])
#Construisons le diagramme en camembert grace à la fonction pie du module pyplot de matplotlib

plt.legend("") #nous n'ajouterons aucune légende en plus

#Les région Nord-est central, Atlantique-sud et Milieu Atlantique sont les régions où les personnes sont les plus heureuses
```

```python
ph = data[(data['happy']=='pretty happy')] 
#Prenons tous les valeurs dont la valeur de la colonne happy est "pretty happy"

x = ph.groupby('region').count()['happy'] 
#Prenons les resultats précedents est regroupons les par région

plt.figure(figsize=(9,6))
#modifions la hauteur et largeur du graphe

plt.pie(x,autopct = lambda x: str(round(x, 2)) + '%', labels = ['Nord-est central', 'Sud-est central', 'Milieu Atlantique', 'Montagnes', 'Nouvelle-Anglettere', 'Pacifique', 'Atlantique-sud', 'Nord-est central', 'Sud-est central'])
#Construisons le diagramme en camembert grace à la fonction pie du module pyplot de matplotlib

plt.legend("") #nous n'ajouterons aucune légende en plus

#Les région citées plus haut sont les régions où la majorité des personnes sont assez heureuses
```

```python
nth = data[(data['happy']=='not too happy')] 
#Prenons tous les valeurs dont la valeur de la colonne happy est "verry happy"

x = nth.groupby('region').count()['happy'] 
#Prenons les resultats précedents est regroupons les par région

plt.figure(figsize=(9,6))
#modifions la hauteur et largeur du graphe

plt.pie(x,autopct = lambda x: str(round(x, 2)) + '%', labels = ['Nord-est central', 'Sud-est central', 'Milieu Atlantique', 'Montagnes', 'Nouvelle-Anglettere', 'Pacifique', 'Atlantique-sud', 'Nord-est central', 'Sud-est central'])
#Construisons le diagramme en camembert grace à la fonction pie du module pyplot de matplotlib

plt.legend("") #nous n'ajouterons aucune légende en plus

plt.legend("")
```

```python
x = data.groupby('region').count()
#regroupons les données du graphe par région

plt.figure(figsize=(15,6))

sns.countplot(x="region", hue="owngun", data=data)
#construisons un histogramme montrant le nombre de personne possédant ou non une arme à feu trié par région

# On voit que les régions où il y a un écart important de valeur sont les régions où le niveau de bonheur est le plus haut (Altantique-sud,milieu atlantique...)
# Contrairement aux régions où l'écart est faible ou la où le taux de personne ayant un port d'arme est supérieur à ceux n'ayant pas
```

```python
x = vh.groupby('region').sum()['tvhours'] #totalisons le nombre heures passé à regarder la télévision par région via la fonction sum
x
# regardons les regions qui totalisent le plus d'heures passé à regarder la télévision
```

```python
plt.figure(figsize=(9,6))
plt.pie(x,autopct = lambda x: str(round(x, 2)) + '%', labels = ['Nord-est central', 'Sud-est central', 'Milieu Atlantique', 'Montagnes', 'Nouvelle-Anglettere', 'Pacifique', 'Atlantique-sud', 'Nord-est central', 'Sud-est central'])
plt.legend("")
# résultat précédent sous forme de diagramme en camembert

#Le résultat semble montrer qu'un haut niveau de bonheur est associé à haut nombre d'heure passé devant la télévision
```
